```python
import numpy as np  
import matplotlib.pyplot as plt 
import math
from scipy.optimize import fsolve
%matplotlib inline
```

Given a function that is bounded by [a,b], we know somewhere in between there, f() evaluated at that point should be zero.

c is the midpoint: (a+b)/(2)

```python
def graph(formula, x_range):  
    x = np.array(x_range)  
    y = formula(x)
    plt.plot(x, y)  
    plt.show()  
    
def my_formula(x):
    return x**2+3*x-np.exp(x**(1/2))

graph(my_formula, range(-10,11))
```

```python
def function1(x):
    return x**2+3*x-np.exp(x**(1/2))
xlist = np.linspace(0,1, num=1000)
plt.figure(figsize=(10, 5))
plt.plot(xlist, function1(xlist), label="Function")
plt.plot(xlist,xlist*0, "--", label="y = 0")
plt.legend()
```

In the x-axis between 0 and 1 there's going to be one root, observe how the function crosses 0

```python
## using -- "fsolve" -- it can find the value numerically

root = fsolve(function1 , 0.5)
print(root)
```

```python
x = np.linspace(-2, 2, 100)
y = x**3 +1/2

def f(x):
    return x**3 +1/2

# set tolerance -- willing to accept as an answer, can't set a function to IFF equals to zero

def bisection(a,b,tol):
    xl = a
    xr = b
    while(np.abs(xl-xr)>= tol):  # as the interval gets smaller, that's where the root is
        c = (xl+xr)/2.0

        prod = f(xl)*f(c)
        if prod > tol: 
            xl = c
        else:
            if prod < tol:
                xr = c
                
    return c

answer = bisection(-5,5,1e-10)

import inspect

z = inspect.getfullargspec(fsolve)

print(z)

ylist = np.linspace(-2,2, num=1000)
plt.plot(x,y)
plt.plot(ylist,ylist*0, "--", label="y = 0")
plt.grid()
    
```
