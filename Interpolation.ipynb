```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

```python
x = [0, 1, 2]
y = [1, 3, 2]

f = interp1d(x, y)
y_hat = f(1.5)
print(y_hat)
```

```python
plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob')
plt.plot(1.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 1.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from scipy.interpolate import interp1d
from scipy.interpolate import interp2d
from scipy.integrate import quad
from scipy.integrate import solve_ivp
```

```python
x_data = np.linspace(0,8,10)
y_data = x_data**4
```

```python
plt.scatter(x_data,y_data)
plt.show()
```

```python
## x axis goes up to meet the corresponding y value
```

```python
plt.plot(x_data, y_data ,'o--')
plt.show()
```

```python
## find y values for many x values
## 'y_f' as y function
```

```python
y_f = interp1d(x_data, y_data, 'linear')
y_f(1.5)

x = np.linspace(0,8,100)
y = y_f(x)

plt.scatter(x,y)
plt.show()
```

```python
## Smooth CUBIC INTERPOLATION
## splines are function approximations that are continous at merging ponits
## matches derivatives at every point, the slopes (1st and 2nd derivative) are the same
## every data point has a piecewise definition of what the best fit is
```

```python
y_f = interp1d(x_data, y_data, 'cubic')

x = np.linspace(0,8,100)
y = y_f(x)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
X = np.arange(11)
Y = np.array([2.0, 1.9, 1.7, 3.5, 0.5, 1.9, 0.8, 0.0, 0.9, 0.2, 2.0])
```

```python
plt.plot(X, Y, "o:")
plt.show()
```

```python
predict = interp1d(X, Y, kind = "nearest") ## x values are rounded off to the nearest integer
X2 = np.linspace(0, 10, 100)
Y2 = np.array([predict(x) for x in X2])

plt.plot(X, Y, "o:")
plt.plot(X2, Y2, "ro:")
plt.legend(["original"])
plt.show()

print(f"X: 1.5, Y: {predict(1.5)}")
```

```python
predict = interp1d(X, Y, kind = "quadratic")
```

```python
X2 = np.linspace(0, 10, 100)
Y2 = np.array([predict(x) for x in X2])
```

```python
plt.plot(X, Y, "o:")
plt.plot(X2, Y2, "ro:")
plt.legend(["original"])
plt.show()
```

```python
print(f"X: 1.5, Y: {predict(1.5)}")
```
