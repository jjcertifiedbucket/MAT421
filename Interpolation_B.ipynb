Lagrange Multiplier for financial plan 

```python
import matplotlib.pyplot as plt
import numpy as np
```

Say we want to begin a start up company, an investor gives a cap budget of 200,000
For this example, I'll focus on two variables:  My employee income and rent for workspace
To make it simple, assume one employee earns 300 a day and one day of renting an office cost 167

```python
employee_pay = 150
cost_workspace = 200
budget = 200000
```

The equation, 300 x (no. of days work) + 167 x (no. of days rented)

The goal is to not spend over the budget of 200000



```python
employee_min = 0
employee_max = budget / employee_pay

cost_workspace_min = 0
cost_workspace_max = budget / cost_workspace
```

```python
def n_employee(n_workspace, budget):
        return (budget - 200 * n_workspace) / 150
```

```python
## if theres a fixed value of rent, then we'll find the remaining budget to pay employees
## same goes the other way
```

```python
def n_workspace(n_employee, budget):
        return (budget - 150 * n_employee) / 200
```

Under the line means it's within the budget, above means more than max.  The goal is to be exactly on the line

```python
emp_x = np.linspace(employee_min, employee_max, 100)
workspace_y = n_workspace(emp_x, budget)

plt.figure(figsize=(8,4))
plt.plot(emp_x, workspace_y)
plt.xlabel('# of days paid to employee')
plt.ylabel('# of days paid for rent')
plt.title('ways of spending the budget')
plt.show()
```

```python
## arbitrary function that measures revenue curve for the business 
## r = 4(no. of days for emp)^3/4 X (no. days for rent)^1/2
```

```python
def revenue(employee, workspace):
    return employee**(3/4) * workspace**(1/2) * 4
```

```python
    fig, (ax_l, ax_r) = plt.subplots(1, 2, figsize = (15, 5))

employee_axis = np.linspace(employee_min, employee_max, 100)
workspace_axis = np.linspace(cost_workspace_min, cost_workspace_max, 100)
employee_grid, workspace_grid = np.meshgrid(employee_axis, workspace_axis)
    
im = ax_l.imshow(revenue(employee_grid, workspace_grid), aspect = 'auto', extent=[employee_min, employee_max, cost_workspace_min, cost_workspace_max])
ax_l.plot(employee_axis, n_workspace(employee_axis, 2500), 'r')
ax_l.set_xlabel('no. of days paid to emp')
ax_l.set_ylabel('no. of days paid for rent')
ax_l.set_title('Possible ways of spending the budget')


employee_axis = np.linspace(employee_min, employee_max)
workspace_axis = np.linspace(cost_workspace_min, cost_workspace_max)
employee_grid, workspace_grid = np.meshgrid(employee_axis, workspace_axis)

im2 = ax_r.contour(revenue(employee_grid,workspace_grid), extent=[employee_min, employee_max, cost_workspace_min, cost_workspace_max])
ax_r.plot(employee_axis, n_workspace(employee_axis, 200000), 'r')
ax_r.set_xlabel('no. of days paid to emp')
ax_r.set_ylabel('no. of days paid for rent')
ax_r.set_title('Possible ways of spending the budget')

plt.colorbar(im,ax=ax_l)
plt.colorbar(im2,ax=ax_r)

plt.show()
```
