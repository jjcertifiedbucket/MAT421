I used a dataset that contains 187 countries that were affected by Covid and which shows the amount of confirmed cases from January 2020 - April 2020.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print('Modules are imported.')
```

```python
corona_dataset_csv = pd.read_csv("Datasets/covid19_Confirmed_dataset.csv")
corona_dataset_csv.head(10)
```

```python
corona_dataset_csv.shape
```

```python
df = corona_dataset_csv.drop(["Lat", "Long"], axis=1)
df.head(10)
```

```python
corona_dataset_csv.drop(["Lat", "Long"], axis=1, inplace=True)
```

```python
corona_dataset_csv.head(10)
```

```python
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()

corona_dataset_aggregated.head(10)
```

```python
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["Italy"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
plt.legend()
```

This is a plot of the infection rates from 01/2020 - 04/2020 from China, Italy, and Spain

```python
corona_dataset_aggregated.describe()
```

Every single day from 01/2020 - 04/2020 are evaluated.  For example, for all 187 countries that were affected by Covid on 1/22/2020, it averaged a mean of 2.96 or approximately 3 confirmed cases among all countries.

The maximum amount of confirmed cases that day 1/22/20 for all countries = 548

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = [] 
for c in countries :
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated['max_infection_rate'] = max_infection_rates ## tab 
```

```python
corona_dataset_aggregated.head()
```

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
corona_data.shape
```
