This is a factorial function (Taylor Series)

```python
def factorial(a):
    result = 1
    for i in range(1, a+1):
        result = result * i
    return result
        
n = input("How many iteration? ")
int_n = int(n)

e = 0.0
for i in range(0, int_n + 1):
    e = e + (1/factorial(i))
    print("iteration -", i, " ", e)
    
```

The higer the iteration the more precise.

```python
import numpy
import math
import matplotlib.pyplot as plt
```

```python
def df(f, x):
    h = 10**-8
    return (f(x+h)-f(x))/h

def example(x):
    return math.sin(x)

def actual(x):
    return math.cos(x)

print(df(example, 2))
print(actual(2))
print(df(example, 2)- actual(2))
```

```python
def f(x):
    return numpy.tan(x)

def f_diff(x):
    return 1/numpy.cos(x)**2
```

```python
def centered (x, h):
    return (f(x+h)-f(x-h))/(2*h)

def forward(x, h):
    return (f(x+h)-f(x))/h
```

```python
x0 = 0.2
```

```python
h_vector = [10**(-temp) for temp in numpy.arange(0, 17, 0.1)]

forward_result = numpy.zeros(len(h_vector))
centered_result = numpy.zeros(len(h_vector))
true_result = numpy.zeros(len(h_vector))
```

```python
for index, i in enumerate(h_vector):
    forward_result[index] = forward(x0, i)
    centered_result[index] = centered(x0, i)
    true_result[index] = f_diff(x0)
```

```python
plt.figure(dpi=100)
plt.loglog(h_vector, abs(forward_result-true_result), label='Forward')
plt.loglog(h_vector, abs(centered_result-true_result), label='Centered')
plt.grid()
plt.xlabel('h')
plt.ylabel('Absolute difference')
plt.legend()
plt.show()
```
